```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd

file_path = '/content/drive/My Drive/Colab Notebooks/adult.csv'
sales_data = pd.read_csv(file_path)
print("First 5 rows of the dataset:")
display(sales_data.head())

display("\nDataset info:")
sales_data.info()

#Read the dataset from the CSV file and clean the data — remove all rows containing the "?" character in any column.
sales_data = sales_data.replace('?', pd.NA)
sales_data = sales_data.dropna()

#Also, verify the correctness of the data (salary and salary in $K).
display(sales_data)

#Display the number of men and women in the dataset.
display("Number of men and women in the dataset:")
display(sales_data['sex'].value_counts())

#Find the average age of men in the dataset.
display("Average age of men in the dataset:")
display(sales_data[sales_data['sex'] == 'Male']['age'].mean())


#Calculate the percentage of people from Poland (native-country).
display("Percentage of people from Poland (native-country):")
display((sales_data[sales_data['native-country'] == 'Poland'].shape[0] / sales_data.shape[0]) * 100)

#Check if there are people without higher education (education: Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters, Doctorate) but earning > 50K.
display("People without higher education but earning > 50K:")
display(sales_data[(sales_data['education'] != 'Bachelors') &
 (sales_data['education'] != 'Prof-school')])

#Obtain age statistics for each education type. Use groupby and describe for this.
display("Age statistics for each education type:")
display(sales_data.groupby('education')['age'].describe())


#Compare the salaries of married and unmarried men. Who earns more? (> 50K or <= 50K) Married men are those whose marital-status starts with Married. Others are considered unmarried.
display("Salaries of married and unmarried men:")
display(sales_data[(sales_data['sex'] == 'Male') & (sales_data['marital-status'].str.startswith('Married'))]['salary'].value_counts())


#Find the maximum number of hours a person works per week. How many people work this many hours per week?
display("Maximum number of hours a person works per week:")
display(sales_data['hours-per-week'].max())



# Read the dataset from the CSV file and clean the data — remove all rows containing the "?" character in any column.
sales_data = sales_data.replace('?', pd.NA)
sales_data = sales_data.dropna()
display(sales_data)


#Calculate the mean and standard deviation of the age for people earning > 50K per year. Then, calculate the same for those earning <= 50K.

sales_data['salary'] = sales_data['salary'].str.strip()

age_stats = sales_data.groupby('salary')['age'].agg(['mean', 'std'])

print("Mean and standard deviation of age:")
print(">50K:")
print("Mean:", age_stats.loc['>50K', 'mean'])
print("Std:", age_stats.loc['>50K', 'std'])

print("\n<=50K:")
print("Mean:", age_stats.loc['<=50K', 'mean'])
print("Std:", age_stats.loc['<=50K', 'std'])




```
