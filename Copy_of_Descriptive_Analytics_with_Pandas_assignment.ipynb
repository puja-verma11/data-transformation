```python
from google.colab import drive
drive.mount('/content/drive')


import pandas as pd

file_path = '/content/drive/My Drive/Colab Notebooks/sales.csv'
sales_data = pd.read_csv(file_path)
print("First 5 rows of the dataset:")
display(sales_data.head())

print("\nDataset info:")
sales_data.info()


# Combine 'Order_Date' and 'Time' into single datetime column and convert to datetime type
sales_data['order_date'] = pd.to_datetime(sales_data['order_date'] + ' ' + sales_data['time'])

# Drop the 'Time' column
sales_data = sales_data.drop(columns=['time'])

# Create new column 'sales_amount' = sales * Quantity
sales_data['sales_amount'] = sales_data['sales'] * sales_data['quantity']

# Transform 'Gender' to 'F' and 'M'
sales_data['gender'] = sales_data['gender'].str.upper().str[0]

# Convert all column names to lowercase
sales_data.columns = sales_data.columns.str.lower()

print("After transformations:")
display(sales_data.head())
sales_data.info()





# 1. Sales Amount summary
max_sales = sales_data['sales_amount'].max()
min_sales = sales_data['sales_amount'].min()
total_sales = sales_data['sales_amount'].sum()
average_sales = sales_data['sales_amount'].mean()
median_sales = sales_data['sales_amount'].median()

sales_summary = f"""
📊 **Sales Amount Summary**

- 🔼 Maximum Sales Amount: ${max_sales:.2f}
- 🔽 Minimum Sales Amount: ${min_sales:.2f}
- 💰 Total Sales: ${total_sales:.2f}
- 📈 Average Sales Amount: ${average_sales:.2f}
- 📊 Median Sales Amount: ${median_sales:.2f}
"""
print(sales_summary)

# 2. Product categories and their popularity
num_categories = sales_data['product'].nunique()
category_counts = sales_data['product'].value_counts()

category_summary = f"""
📦 **Product Category Summary**

- 🛍️ The company sells **{num_categories}** unique products.

🔢 **Popularity of Each Product (Top 10):**
"""
for product, count in category_counts.head(10).items():
    category_summary += f"- {product}: {count} orders\n"

print(category_summary)

# 3. Top 5 most popular products
top_5_products = category_counts.head(5)
top_5_summary = "🏆 **Top 5 Most Popular Products:**\n\n"
for i, (product, count) in enumerate(top_5_products.items(), start=1):
    top_5_summary += f"{i}. {product}: {count} orders\n"
print(top_5_summary)


# 4. Who purchases more often, women or men?
gender_counts = sales_data['gender'].value_counts()

gender_summary = f"""
👥 **Purchase Frequency by Gender:**

- Women (F): {gender_counts.get('F', 0)} orders
- Men (M): {gender_counts.get('M', 0)} orders
"""

if gender_counts.get('F', 0) > gender_counts.get('M', 0):
    gender_summary += "\n➡️ Women purchase more often than men."
elif gender_counts.get('M', 0) > gender_counts.get('F', 0):
    gender_summary += "\n➡️ Men purchase more often than women."
else:
    gender_summary += "\n➡️ Both genders purchase equally often."

print(gender_summary)


# 5. Most popular payment method
payment_counts = sales_data['payment_method'].value_counts()
most_popular_payment = payment_counts.idxmax()
payment_summary = f"""
💳 **Most Popular Payment Method:**

- {most_popular_payment} with {payment_counts.max()} orders
"""

print(payment_summary)
```
